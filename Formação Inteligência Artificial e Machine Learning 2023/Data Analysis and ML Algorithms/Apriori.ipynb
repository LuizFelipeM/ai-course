```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
with open('Transacoes.txt', 'r') as f:
  transactions = [line.strip().split(',') for line in f.readlines()]

transactions
```

```python
te = TransactionEncoder()
te_arr = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_arr, columns=te.columns_)
df
```

```python
frequency_itemsets = apriori(df, min_support=0.5, use_colnames=True)
frequency_itemsets
```

```python
rules = association_rules(frequency_itemsets, metric='confidence', min_threshold=0.5)
rules
```
