```python
import transformers
from transformers import pipeline
```

```python
mask = pipeline("fill-mask", model="neuralmind/bert-base-portuguese-cased")
text = mask("Existe uma chance do corpo cair no [MASK]")
for x in range(len(text)):
  print(text[x])
```

```python
text = mask(["Existe uma chance do corpo cair no [MASK]", "Brasília é a capital do [MASK]"])
for x in range(len(text)):
  print(text[x])
```
