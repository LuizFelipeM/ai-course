```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
np.random.seed(5)
data = np.random.normal(0, 1, 100) + np.linspace(0, 1, 100)
data[50] = 4
```

```python
model = SimpleExpSmoothing(data)
model_fit = model.fit(smoothing_level=0.2)
```

```python
data_pred = model_fit.predict(0, len(data) - 1)
data_pred
```

```python
anomalies = np.where(np.abs(data - data_pred) > 3 * np.std(data))[0]
anomalies
```

```python
data[anomalies]
```

```python
plt.plot(data, label="Original data")
plt.plot(data_pred, label="Exponential smoothing")
plt.plot(anomalies, data[anomalies], 'ro', label="Anomalies")
plt.legend()
plt.show()
```
