```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(0, 1, 100) + np.linspace(0, 1, 100)
data[50] = 7
data[25] = 6
```

```python
plt.plot(data, label="Data")
plt.title("Original data")
plt.show()
```

```python
s = pd.Series(data)
s_ma = s.rolling(10).mean()
```

```python
plt.plot(s_ma, label="Data")
plt.title("Moving average data")
plt.show()
```

```python
anomalies = s[np.abs(s - s_ma) > 3 * s.std()]
anomalies
```

```python
plt.plot(s, label="Data")
plt.plot(s_ma, label="Moving average")
plt.plot(anomalies, 'ro', label="Anomalies")
plt.title("Data")
plt.show()
```
