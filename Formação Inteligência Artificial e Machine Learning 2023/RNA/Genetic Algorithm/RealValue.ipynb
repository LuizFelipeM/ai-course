```python
import pygad
import numpy as np
```

```python
# Equation 2*x + 5 = 20
def fitness(ga_instance, solution, solution_idx):
  x = solution[0]
  result = 2*x + 5
  expected_result = 20
  fitness_value = 1.0 / np.abs(result - expected_result + 1e-8)
  return fitness_value
```

```python
ga_instance = pygad.GA(num_generations=50,
                       num_parents_mating=10,
                       fitness_func=fitness,
                       sol_per_pop=20,
                       num_genes=1,
                       gene_type=float,
                       init_range_low=-10,
                       init_range_high=10)

ga_instance.run()
solution, solution_fitness, _ = ga_instance.best_solution()
print(f"Best solution: {round(solution[0], 1)}")
```
